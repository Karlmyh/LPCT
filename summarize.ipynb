```python
# calculate rank sum
def rank_sum_per_row(arr):
    num_rows, num_cols = arr.shape
    rank_sums = []

    for i in range(num_rows):
        row = - arr[i, :]
        # Calculate the ranks of elements in the row
        ranks = np.argsort(row).argsort() + 1
        rank_sums.append(ranks)

    return np.array(rank_sums)

summary.replace(np.nan, 0, inplace = True)
npsummary = summary.to_numpy()



rank_sum_per_row(npsummary[:,1:]).sum(axis = 0)
```

```python
# find the three best methods
data_file_name_seq = [
"anonymity",
# "census",
# "diabetes",
# "election",
# "email",
# "employ",
# "employee",
# "jobs",
# "rice",
# "landcover",
# "taxidata",
]
sig_level = 0.05
specific_epsilon = 8
method1 = "LPCT-V"
method2 = "LPCT-M"



report_method = [method1, method2]


all_log = pd.DataFrame([])
log_file_dir = "./results/realdata" 
for method in report_method:
    log = pd.read_csv("{}/{}.csv".format(log_file_dir,method), header=None)
    log.columns = "dataname,method,iterate,epsilon,n_P,n_Q,acc,bce,time,parameter1,parameter2,parameter3".split(',')
    all_log = all_log.append(log)
    


######## filter the results to display
summarize_log = all_log.copy()
summarize_log = summarize_log[summarize_log["dataname"].isin(data_file_name_seq)]
summarize_log = summarize_log[summarize_log["method"].isin(report_method)]
summarize_log = summarize_log[summarize_log["epsilon"] == specific_epsilon]


mean_result_log = summarize_log.groupby(by=["dataname", "method", "epsilon", "parameter1", "parameter2", "parameter3"]).agg('mean').reset_index()
# find the three best methods
data_file_name_seq = [
# "anonymity",
"census",
# "diabetes",
# "election",
# "email",
# "employ",
# "employee",
# "jobs",
# "rice",
# "landcover",
# "taxidata",
]
sig_level = 0.05
specific_epsilon = 8
method1 = "LPCT-V"
method2 = "LPCT-M"



report_method = [method1, method2]


all_log = pd.DataFrame([])
log_file_dir = "./results/realdata" 
for method in report_method:
    log = pd.read_csv("{}/{}.csv".format(log_file_dir,method), header=None)
    log.columns = "dataname,method,iterate,epsilon,n_P,n_Q,acc,bce,time,parameter1,parameter2,parameter3".split(',')
    all_log = all_log.append(log)
    


######## filter the results to display
summarize_log = all_log.copy()
summarize_log = summarize_log[summarize_log["dataname"].isin(data_file_name_seq)]
summarize_log = summarize_log[summarize_log["method"].isin(report_method)]
summarize_log = summarize_log[summarize_log["epsilon"] == specific_epsilon]


mean_result_log = summarize_log.groupby(by=["dataname", "method", "epsilon", "parameter1", "parameter2", "parameter3"]).agg('mean').reset_index()


maxidx1 = mean_result_log[mean_result_log["method"].isin([method1])]["acc"].argmax()
dic1 = {
    "epsilon":[specific_epsilon],
    "dataname":[dataname],
    "method": [method1],
    "parameter1": [mean_result_log.loc[maxidx1]["parameter1"]],
    "parameter2": [mean_result_log.loc[maxidx1]["parameter2"]],
    "parameter3": [mean_result_log.loc[maxidx1]["parameter3"]]
}
mask_list = [all_log[col].isin(values) for col, values in dic1.items()]
combined_mask = mask_list[0]
for mask in mask_list[1:]:
    combined_mask = combined_mask & mask
filtered_log = all_log[combined_mask]
seq1 = filtered_log["acc"]


maxidx2 = mean_result_log[mean_result_log["method"].isin([method2])]["acc"].argmax()
dic2 = {
    "epsilon":[specific_epsilon],
    "dataname":[dataname],
    "method": [method2],
    "parameter1": [mean_result_log.loc[maxidx2]["parameter1"]],
    "parameter2": [mean_result_log.loc[maxidx2]["parameter2"]],
    "parameter3": [mean_result_log.loc[maxidx2]["parameter3"]]
}
mask_list = [all_log[col].isin(values) for col, values in dic2.items()]
combined_mask = mask_list[0]
for mask in mask_list[1:]:
    combined_mask = combined_mask & mask
filtered_log = all_log[combined_mask]
seq2 = filtered_log["acc"]


wilcoxon( seq1 - seq2, alternative = "less").pvalue < sig_level
```
